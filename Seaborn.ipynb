```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv')
df
```

# Lineplot

```python
plt.figure(figsize=(16,9))
sns.set(style='white')
sns.lineplot(x=df['year'],y=df['passengers'],data=df,dashes=True,hue='month',style='month',
             legend='full')
plt.title('SEA_Born')
plt.xlabel('yea')
plt.legend(loc='lower right')
plt.grid(True,alpha=0.5,linewidth=0.8,linestyle='--',color='r')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='white')
sns.barplot(x=tips.day,y=tips.total_bill,data=tips,order=(['Sun','Thur','Fri','Sat']),hue='sex',
           saturation=10,palette='hot',alpha=0.7,linestyle='-.',linewidth=5,edgecolor='k',
            
            hue_order=['Male','Female'],estimator=np.sum,ci=50,errcolor='y',errwidth=5,
            hatch='+',
            
           )
# plt.title('Total_Bill Vs Day',fontdict={'fontsize':20,'fontweight':'bold'})
plt.ylabel('Total_Bill (Sum)')
plt.title('Total_Bill Vs Day',fontsize=20,fontweight='bold')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='white')
a={"alpha":0.5,"linestyle":'-.',"linewidth":5,"palette":'summer'}
sns.barplot(y=tips.day,x=tips.total_bill,data=tips,order=(['Sun','Thur','Fri','Sat']),hue='sex',
           saturation=10,edgecolor='k',
            
            hue_order=['Male','Female'],estimator=np.sum,ci=50,errcolor='y',errwidth=5,**a
            
           )
# plt.title('Total_Bill Vs Day',fontdict={'fontsize':20,'fontweight':'bold'})
plt.ylabel('Total bill (sum)')
plt.title('Total_Bill Vs Day',fontsize=20,fontweight='bold')
plt.grid(True)
plt.show()
```
